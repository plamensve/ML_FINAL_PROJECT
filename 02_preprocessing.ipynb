```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('main_data_source/diabetes_dataset.csv')
data
```

```python
data['gender'] = data['gender'].map({'Male': 0, 'Female': 1})
data
```

```python
data['ethnicity'].value_counts()
```

```python
ethnicity_dummies = pd.get_dummies(data['ethnicity'], prefix='ethnicity').astype(int)
data = pd.concat([data, ethnicity_dummies], axis=1)
data = data.drop(columns='ethnicity')
data
```

```python
data['education_level'].value_counts()
```

```python
education_level_dummies = pd.get_dummies(data['education_level'], prefix='education_level').astype(int)
data = pd.concat([data, education_level_dummies], axis=1)
data = data.drop(columns='education_level')
data = data.rename(columns={'education_level_No formal': 'education_level_no_formal'})
data
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('main_data_source/diabetes_dataset.csv')
data
```

```python
data['gender'] = data['gender'].map({'Male': 0, 'Female': 1})
data
```

```python
data['ethnicity'].value_counts()
```

```python
ethnicity_dummies = pd.get_dummies(data['ethnicity'], prefix='ethnicity').astype(int)
data = pd.concat([data, ethnicity_dummies], axis=1)
data = data.drop(columns='ethnicity')
data
```

```python
data['education_level'].value_counts()
```

```python
education_level_dummies = pd.get_dummies(data['education_level'], prefix='education_level').astype(int)
data = pd.concat([data, education_level_dummies], axis=1)
data = data.drop(columns='education_level')
data = data.rename(columns={'education_level_No formal': 'education_level_no_formal'})
data
```

```python
data['income_level'].value_counts()
```

```python
income_level_dummies = pd.get_dummies(data['income_level'], prefix='income_level').astype(int)
data = pd.concat([data, income_level_dummies], axis=1)
data = data.drop(columns='income_level')
data
```

```python
data['employment_status'].value_counts()
```

```python
employment_status_dummies = pd.get_dummies(data['employment_status'], prefix='employment_status').astype(int)
data = pd.concat([data, employment_status_dummies], axis=1)
data = data.drop(columns='employment_status')
data
```

```python
data['smoking_status'].value_counts()
```

```python
smoking_status_dummies = pd.get_dummies(data['smoking_status'], prefix='smoking_status').astype(int)
data = pd.concat([data, smoking_status_dummies], axis=1)
data = data.drop(columns='smoking_status')
data
```

```python
data['diabetes_stage'].value_counts()
```

```python
diabetes_stage_dummies = pd.get_dummies(data['diabetes_stage'], prefix='diabetes_stage').astype(int)
data = pd.concat([data, diabetes_stage_dummies], axis=1)
data = data.drop(columns='diabetes_stage')
data
```

```python
data = data.rename(columns={'diagnosed_diabetes': 'target'})
data.columns = data.columns.map(lambda x: x.strip().lower().replace(' ', '_').replace('-', '_'))
data
```

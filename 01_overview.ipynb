### Body mass index [BMI]
Description:
bmi (Body Mass Index) is a numerical value that measures body fat based on weight and height. It is commonly used to assess whether a person’s weight is within a healthy range.

$$
BMI = \frac{weight\ (kg)}{height\ (m)^2}
$$

Example use:
Frequently used in health data analysis and classification tasks to study relationships between body composition and risk factors such as heart disease, diabetes, or hypertension.

```python
data[['bmi']].describe().T
```

```python
plt.figure(figsize=(7, 5))
plt.hist(data['bmi'], edgecolor='black')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.show()
```

### Systolic Blood Pressure
Description:
systolic_bp represents the systolic blood pressure, which is the pressure in the arteries when the heart contracts and pumps blood through the body.

Values:

Normal: < 120 mmHg

Elevated: 120 – 129 mmHg

High (Hypertension Stage 1): 130 – 139 mmHg

High (Hypertension Stage 2): ≥ 140 mmHg

Example use:
Used as a key health indicator or feature in medical and health-related datasets to assess cardiovascular risk, predict hypertension, or analyze the impact of lifestyle factors such as diet and exercise.

```python
data[['systolic_bp']].describe().T
```

```python
plt.figure(figsize=(7, 5))
plt.hist(data['systolic_bp'], edgecolor='black')
plt.title('Systolic Blood pressure Distribution')
plt.xlabel('SBP')
plt.ylabel('Count')
plt.show()
```

### Diastolic Blood pressure
Description:
diastolic_bp refers to the diastolic blood pressure, which measures the pressure in the arteries when the heart is resting between beats.

Values:

Normal: < 80 mmHg

Elevated: 80 – 89 mmHg

High (Hypertension Stage 2): ≥ 90 mmHg

Example use:
Commonly used as a clinical indicator in cardiovascular studies and health datasets to evaluate blood pressure levels, monitor hypertension, and analyze the relationship between lifestyle, stress, and heart health.

### Heart Rate

Example use:
Used as a vital sign or feature in health datasets to monitor cardiac activity, detect irregularities such as arrhythmia, and study correlations between heart rate, stress levels, and physical fitness.


### Cholesterol total
Description:
cholesterol_total represents the total amount of cholesterol in the blood, measured in milligrams per deciliter (mg/dL). It includes low-density lipoprotein (LDL), high-density lipoprotein (HDL), and a portion of triglycerides.

Values:

Desirable: < 200 mg/dL

Borderline high: 200 – 239 mg/dL

High: ≥ 240 mg/dL

Example use:
Serves as a key cardiovascular health indicator in medical datasets, often used in classification or regression tasks to assess heart disease risk, analyze lipid profiles, or study the effects of diet and lifestyle.

### HDL Cholesterol
Description:
hdl_cholesterol measures the “good” cholesterol, which helps remove excess cholesterol from the bloodstream and reduces the risk of heart disease.

Values:

Normal: 40–59 mg/dL

High (protective): ≥ 60 mg/dL

Example use:
Used as a predictive feature in health and cardiovascular datasets to evaluate lipid balance, model heart disease risk, or study the impact of diet and physical activity on cholesterol levels.

### Triglycerides

Normal: < 150 mg/dL

High: 200 – 499 mg/dL

Very high: ≥ 500 mg/dL

Example use:
Used as a key biochemical feature in health and metabolic datasets to study cardiovascular risk, lipid metabolism, and the impact of diet, obesity, or diabetes on blood fat levels.

### Glucose fasting

### Glucose postprandial
Description:
glucose_postprandial represents the blood glucose level measured after eating (typically 2 hours after a meal), expressed in milligrams per deciliter (mg/dL). It reflects how the body manages sugar intake and insulin response after food consumption.

Values:

Normal: < 140 mg/dL

Prediabetes: 140 – 199 mg/dL

Diabetes: ≥ 200 mg/dL

Example use:
Used in metabolic and endocrinology datasets to evaluate glucose tolerance, predict diabetes onset, and analyze post-meal blood sugar fluctuations in relation to diet and insulin sensitivity.

```python
data[['glucose_postprandial']].describe().T
```

### Insulin level
Description:
insulin_level indicates the amount of insulin present in the blood, measured in micro–units per milliliter (µU/mL). It helps assess how effectively the body regulates blood glucose and responds to insulin.

Values:

Normal (fasting): 2 – 25 µU/mL

Elevated: > 25 µU/mL (may indicate insulin resistance or hyperinsulinemia)

Example use:
Commonly used in metabolic and diabetes-related datasets to study insulin resistance, classify metabolic syndrome, or predict type 2 diabetes risk based on glucose–insulin interaction patterns.

```python
data[['insulin_level']].describe().T
```

```python
plt.figure(figsize=(7, 5))
plt.hist(data['insulin_level'], edgecolor='black')
plt.title('Insulin level Distribution')
plt.xlabel('Insulin level')
plt.ylabel('Count')
plt.show()
```

### Hemoglobin A1c
Description:
hba1c measures the average blood glucose level over the past 2–3 months, expressed as a percentage. It reflects the proportion of hemoglobin that has glucose attached to it and is a key indicator of long-term blood sugar control.

Values:

Normal: < 5.7%

Prediabetes: 5.7% – 6.4%

Diabetes: ≥ 6.5%

Example use:
Used extensively in diabetes monitoring and prediction models to assess glycemic control, identify chronic hyperglycemia, and evaluate the effectiveness of treatment or lifestyle interventions.

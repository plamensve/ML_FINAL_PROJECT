```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import  KMeans
```

```python
blobs_data, blobs_clusters = make_blobs(n_samples=500, centers=4, random_state=42)
blobs_data
```

```python
k_means = KMeans(
    n_clusters=4,
    init='random',
)

k_means.fit(
    blobs_data
)
```

```python
k_means.cluster_centers_
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(blobs_data[:, 0], blobs_data[:, 1], c = blobs_clusters, s=30)
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], color='g')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
